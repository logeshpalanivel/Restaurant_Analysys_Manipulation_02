```python
top_10_descriptions = df['ITEM_NAME'].value_counts().nlargest(10)          # without qty
sns.barplot(x=top_10_descriptions,y=top_10_descriptions.index)             # This is used for check how mny times the customer visited
plt.xlabel('Count')
plt.ylabel('Product')
plt.title('Top 10 Product Descriptions')
plt.show()
```

```python
top_10 = df.groupby('ITEM_NAME')['Qty.'].sum().nlargest(10)

sns.barplot(x=top_10.values,y=top_10.index)
plt.xlabel('Count')
plt.ylabel('Product')
plt.title('Top_10')
plt.xticks(rotation=50)
plt.show()
```

```python
top_10
```

```python
top10_A = df.groupby('ITEM_NAME')['FINAL_TOTAL'].sum().nlargest(10)
sns.barplot(x=top10_A.values,y=top10_A.index)
plt.xlabel('Count')
plt.ylabel('Product')
plt.title('Top10')
plt.xticks(rotation=50)
plt.show()
```

```python
top10_A
```

```python
top10_d = df.groupby('ITEM_NAME')['DISCOUNT'].sum().nlargest(10)
sns.barplot(x=top10_d.values,y=top10_d.index)
plt.xlabel('Count')
plt.ylabel('product')
plt.title('top10_d')
plt.show()
```

```python
top10_tax = df.groupby('ITEM_NAME')['TAX'].sum().nlargest(10)
sns.barplot(x=top10_tax.values,y=top10_tax.index)
plt.xlabel('Count')
plt.ylabel('product')
plt.title('top10_tax')
plt.show()
```
